```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn import tree
```

```python
path=r"Company_Data.csv"
df1=pd.read_csv(path)
df1
```

```python
df1.loc[df1['Sales'] > 10, ['Sales_df']]='high'
```

```python
df1.loc[df1['Sales'] < 10, ['Sales_df']]='low'
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
categorical_columns=df1.select_dtypes(include=['object']).columns
```

```python
categorical_columns
```

```python
df1['ShelveLoc']=label.fit_transform(df1['ShelveLoc'])
```

```python
df1['Urban']=label.fit_transform(df1['Urban'])
```

```python
df1['US']=label.fit_transform(df1['US'])
```

```python
df1.Sales_df.value_counts()
```

```python
df1['Sales_df'].dtypes
```

```python
sales_df=pd.get_dummies(df1['Sales_df'],drop_first=True,prefix='sales_df')
```

```python
sales_df
```

```python
df1=df1.drop('Sales_df',axis=1)
df1
```

```python
df1=pd.concat([df1,sales_df],axis=1)
```

```python
df1
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
X=df1.iloc[:,1:-1]
```

```python
y=df1.iloc[:,-1]
```

```python
print(X)
```

```python
print(y)
```

```python
kfold=KFold(n_splits=10)
```

```python
cart=DecisionTreeClassifier()
```

```python
num_tress=100
```

```python
model=BaggingClassifier(base_estimator=cart,n_estimators=num_tress)
```

```python
results=cross_val_score(model,X,y,cv=kfold)
```

```python
print(results.mean())
```

# Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1=RandomForestClassifier(n_estimators=num_tress,max_features=3,)
```

```python
results=cross_val_score(model1,X,y,cv=kfold)
```

```python
print(results.mean())
```

# AdaBoost Classification

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model2=AdaBoostClassifier(n_estimators=10,random_state=7)
```

```python
results=cross_val_score(model2,X,y,cv=kfold)
```

```python
print(results.mean())
```

# Stacking

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
```

```python
kfold=KFold(n_splits=10,random_state=8)
```

```python
estimators=[]
model3=LogisticRegression(max_iter=500)
estimators.append(('logistic',model3))
```

```python
model4=DecisionTreeClassifier()
```

```python
estimators.append(('cart',model4))
```

```python
model5=SVC()
```

```python
estimators.append(('svm',model5))
```

```python
ensemble=VotingClassifier(estimators)
```

```python
results=cross_val_score(ensemble,X,y,cv=kfold)
```

```python
print(results.mean())
```
